#### 2. Now let's compute the duration variable. It should contain the duration of a ride in minutes.

#### What's the standard deviation of the trips duration in January?

```python
df_january.head()
```

```python
df_january['duration'] = df_january['tpep_dropoff_datetime'] - df_january['tpep_pickup_datetime']
```

```python
df_january['duration_mins'] = df_january['duration'].dt.total_seconds() / 60
```

```python
df_january.head()
```

```python
# Standard deviation of the trips duration
df_january.duration_mins.std()
# 42.59
```
